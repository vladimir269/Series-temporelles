# **Projet de Série temporelles: Pollution à Pékin**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
data = pd.read_csv('air_pollution.csv', index_col='date', parse_dates=True)
poll = data['pollution_today']
```

```python
decompose_mult = seasonal_decompose(poll, model='multiplicative', period=None)
decompose_add = seasonal_decompose(poll, model='additive', period=None)

tendance_mult = decompose_mult.trend
saisonnalite_mult = decompose_mult.seasonal
residus_mult = decompose_mult.resid

tendance_add = decompose_add.trend
saisonnalite_add = decompose_add.seasonal
residus_add = decompose_add.resid
```

```python
plt.figure(figsize=(15, 15))
plt.subplot(711)
plt.plot(poll, label='Données brutes')
plt.legend(loc='upper left')

plt.subplot(712)
plt.plot(tendance_mult, label='Tendance modèle multiplicatif')
plt.legend(loc='upper left')

plt.subplot(713)
plt.plot(tendance_add, label='Tendance modèle additif')
plt.legend(loc='upper left')

plt.subplot(714)
plt.plot(saisonnalite_mult[0:50], label='Saisonnalité modèle multiplicatif')
plt.legend(loc='upper left')

plt.subplot(715)
plt.plot(saisonnalite_add[0:50], label='Saisonnalité modèle additif')
plt.legend(loc='upper left')

plt.subplot(716)
plt.plot(residus_mult, label='Résidus (Stationnaires) modèle multiplicatif')
plt.legend(loc='upper left')

plt.subplot(717)
plt.plot(residus_add, label='Résidus (Stationnaires) modèle additif')
plt.legend(loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(15, 15))

plt.subplot(411)
plt.plot(poll, label='Données brutes')
plt.legend(loc='upper left')

plt.subplot(412)
plt.plot(poll.diff().dropna(), label='diff')
plt.legend(loc='upper left')

plt.subplot(413)
plt.plot(np.log(poll), label='log')
plt.legend(loc='upper left')

plt.subplot(414)
plt.plot(np.log(poll).diff().dropna(), label='log+diff')
plt.legend(loc='upper left')
plt.show()
```
